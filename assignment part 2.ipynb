```python
import numpy as np
import pandas as pd

# 1. Lists
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]

# 2. Arrays
array1 = np.array(list1)
array2 = np.array(list2)

# 3. Identify their type using type()
print("Type of array1:", type(array1))
print("Type of array2:", type(array2))

# 4. Perform mathematical operations
print("Multiplication:", array1 * array2)
print("Division:", array1 / array2)
print("Power off:", array1 ** array2)

# 5. Combine text with Numpy function
print("Addition of Two:", array1, "+", array2)

# 6. Implement np.sin(), log(), log2(), np.exp()
print("sin(array1):", np.sin(array1))
print("log(array1):", np.log(array1))
print("log2(array1):", np.log2(array1))
print("exp(array1):", np.exp(array1))

import matplotlib.pyplot as plt

# Generate data for healthcare
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40, 45]  # Extra value added to age for alignment
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.scatter(height, weight)
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()

# Bar Chart
plt.bar(people, height)
plt.title('Bar Chart of People vs Height')
plt.xlabel('People')
plt.ylabel('Height')
plt.show()

# Histogram
plt.hist(age, bins=5)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# You can use pandas to read data from various sources

# 1. Access your own dataset using Google Drive in Google Colab
# For example:
# data = pd.read_excel("/content/drive/MyDrive/your_file.xlsx")

# 2. Upload the text file using tab separated value(.tsv)
# For example:
# data = pd.read_csv("your_file.tsv", sep="\t")

# 3. Upload the text file using comma separated value(.csv)
# For example:
# data = pd.read_csv("your_file.csv")

# 4. Access the excel file using .xlsx
# For example:
# data = pd.read_excel("your_file.xlsx")

# 5. Access the text from the URL
# For example:
# url = "https://example.com/your_file.txt"
# data = pd.read_csv(url)


```
